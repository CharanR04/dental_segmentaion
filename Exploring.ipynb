```python
import cv2
from torchvision import transforms as T
from torch.utils.data import DataLoader
from helper.Dataset import CustomDataset
from helper.Dataset import img_helper
import matplotlib.pyplot as plt
```

```python
img_helper.genreate_mask('Base')
```

```python
img = cv2.imread('Data/Train/Base/2.jpg')
img_hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
img_hls = cv2.cvtColor(img, cv2.COLOR_BGR2HLS)
mask = cv2.imread('Data/Train/White/Mask/2.jpg')
```

```python
fig,ax = plt.subplots(1,4, figsize = (20,20))
h = img_hsv[:,:,0]
s = img_hsv[:,:,1]
v = img_hsv[:,:,2]
ax[0].set_title('Hue')
ax[0].imshow(h,cmap="gray")
ax[1].set_title('Saturation')
ax[1].imshow(s,cmap="gray")
ax[2].set_title('Value')
ax[2].imshow(v,cmap="gray")
ax[3].set_title('Mask')
ax[3].imshow(mask)
```

```python
fig,ax = plt.subplots(1,4, figsize = (20,20))
h = img_rgb[:,:,0]
s = img_rgb[:,:,1]
v = img_rgb[:,:,2]
ax[0].set_title('Red')
ax[0].imshow(h, cmap="gray")
ax[1].set_title('Green')
ax[1].imshow(s,cmap="gray")
ax[2].set_title('Blue')
ax[2].imshow(v,cmap="gray")
ax[3].set_title('Mask')
ax[3].imshow(mask)
```

```python
fig,ax = plt.subplots(1,4, figsize = (20,20))
h = img_hls[:,:,0]
l = img_hls[:,:,1]
s = img_hls[:,:,2]
ax[0].set_title('Hue')
ax[0].imshow(h,cmap="gray")
ax[1].set_title('Saturation')
ax[1].imshow(l,cmap="gray")
ax[2].set_title('Value')
ax[2].imshow(s,cmap="gray")
ax[3].set_title('Mask')
ax[3].imshow(mask)
```

```python
transforms = T.Compose([T.ToTensor(),T.CenterCrop(748)])
dataset = CustomDataset(root_dir = 'Data/Train/White', transforms = transforms)
data_loader = DataLoader(dataset, batch_size= 5, shuffle = True)
data = next(iter(data_loader))
```
